```python
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
import datetime as dt
import numpy as np
```

```python
df = pd.read_csv('us_counties_covid19_daily.csv', parse_dates=['date'], dayfirst=True)
```

```python
df.head()
```

```python
df = df[['date', 'state','cases','deaths']]
```

```python
df.head()
```

```python
today = df[df.date == '2020-05-09']
```

```python
today
```

```python
max_confirmed_cases=today.sort_values(by="deaths",ascending=False)
max_confirmed_cases
```

```python
top_states_confirmed=max_confirmed_cases[0:5]

sns.set(rc={'figure.figsize':(10,10)})
sns.barplot(x="state",y="deaths", data=top_states_confirmed, hue="state")
plt.show()
```

```python
max_death_cases=today.sort_values(by="cases",ascending=False)
max_death_cases
```

```python
top_states_death=max_death_cases[0:5]

sns.set(rc={'figure.figsize':(10,10)})
sns.barplot(x="state",y="deaths",data=top_states_death,hue="state")
plt.show()
```

```python
newY = df[df.state == 'New York']
```

```python
newY
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date", y="deaths",data = newY, color="r")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date", y="cases",data = newY, color="b")
plt.show()
```
